```python
products_info = pd.read_csv(data_path + 'products_info.csv', index_col=0)
```

```python
index_list = []
for i in products_info.index:
    index_list.append(i)
products_info = products_info.reset_index()
products_info['LP ID'] = index_list
```

```python
products_info.head()
```

```python
products_info[['Sector_1', 'Sector_2', "Sector_3"]] = products_info['Sector(s)'].str.split(';', expand=True)
products_info[['Main_Category', 'Sub_Category']] = products_info['Primary Essential Function'].str.split('-',1, expand=True)
products_info = products_info.drop(columns=["Sector(s)","Primary Essential Function","URL"])
```

```python
products_info["LP ID"] = products_info["LP ID"].astype(float)
```

```python
districts = pd.read_csv(data_path + 'districts_info.csv', index_col=0)
```

```python
districts.isnull().sum()
```

```python
districts = districts.dropna(subset = ['state'])# as some of the state data is not available.
```

```python
index_list = []
for i in districts.index:
    index_list.append(i)
districts = districts.reset_index()
districts['district_id'] = index_list
```

```python
districts.head()
```

```python
districts.county_connections_ratio.value_counts()
```

```python
districts["pct_free/reduced"].value_counts()
```

```python
i = districts[districts["county_connections_ratio"]=="[1, 2["].index
districts=districts.drop(i)
```

```python
districts.isnull().sum()
```

```python
for i in districts.columns[2:]:
    districts[i] = districts[i].str.strip('[]') #Se eliminan los corchetes
districts.head()
```

```python
# Obteniendo los nombres de los archivos
globbed_files = glob.glob(data_path + 'engagement_data/*.csv')
engagement = []
```

```python
for filename in globbed_files:
    frame = pd.read_csv(filename)
    frame['district_id'] = os.path.basename(filename)
    engagement.append(frame)

# Concatenando todos los datos en un dataframe
engagement = pd.concat(engagement, ignore_index=True)
engagement['district_id'] = engagement['district_id'].str.strip('.csv').astype(int)
```

```python
engagement['time'] = pd.to_datetime(engagement['time'], errors = 'coerce')
engagement["Month"] = pd.DatetimeIndex(engagement['time']).month
engagement["Year"] = pd.DatetimeIndex(engagement['time']).year
```

```python
engagement.isnull().sum() #Hay muchos valores nulos en engagement_index.
```

```python
engagement = engagement.dropna(subset = ['engagement_index',"lp_id"]) #Se borran los valores nulos.
```

```python
# Ordenando las tablas
engage_districts = pd.merge(engagement,districts, how="inner",left_on="district_id" , right_on="district_id")
```

```python
Final_dataframe = pd.merge(engage_districts,products_info, how="inner",left_on="lp_id" , right_on="LP ID")
```

```python
engage_table = Final_dataframe.sample(frac=0.50) #Se muestrean los datos a solo el 50% ya que tenemos demasiadas filas.
```

**1. ¿Cuál es el efecto de la pandemia de COVID-19 en el aprendizaje en línea y a distancia, y cómo podría evolucionar esto también en el futuro?**

```python
engage_table["engagement_index"] = engage_table["engagement_index"].astype(float)
```

```python
Monthwise = engage_table.groupby("Month")["engagement_index"].mean()
plt.figure(figsize = (12,6))
sns.pointplot(x=Monthwise.index, y=Monthwise.values,color='magenta')
plt.xlabel('Meses',fontsize = 10,color='darkblue')
plt.ylabel('Indice Medio de Participacion',fontsize = 10,color='darkblue')
plt.title('Indice de Participacion Mensual en 2020',fontsize = 15,color='darkblue')
```

**2. ¿Cómo cambia la participación de los estudiantes con los diferentes tipos de tecnología educativa durante el transcurso de la pandemia?**

```python
edtech=engage_table.groupby(['Month','Product Name'])['engagement_index'].mean().unstack()
top10edtech=edtech.sum(axis=0).sort_values(ascending=False).head(10)#Comparando solo los 20 mejores, ya que tenemos muchos productos.
```

```python
dftop10edtech = edtech[top10edtech.index].copy()
dftop10edtech["Month"] = dftop10edtech.index
dftop10edtech = dftop10edtech.melt('Month', var_name='cols',  value_name='vals')

plt.figure(figsize = (12,8))
g = sns.catplot(x="Month", y="vals", hue='cols', data=dftop10edtech, kind='point')
plt.xlabel('Meses',fontsize = 10,color='darkblue')
plt.ylabel('Indice de Participacion',fontsize = 10,color='darkblue')
plt.title('Indice de Participacion del Top 10 en 2020',fontsize = 12,color='darkblue')
```

**3. ¿Cómo se relaciona la participación de los estudiantes con las plataformas de aprendizaje en línea con las diferentes geografías? ¿Contexto demográfico (p. ej., raza/etnicidad, ESL, discapacidad de aprendizaje)? ¿Contexto de aprendizaje? Estatus socioeconómico**

```python
# Creando un grafico de barras
state_ei = engage_table.groupby(by=["state"])['engagement_index'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,4),dpi=100)
plt.title(' Indice de Participacion Medio Estatal')
sns.barplot(x=state_ei.index,y=state_ei.values, palette = "viridis",edgecolor='black')
plt.ylabel("Indice Medio de Participacion", fontsize=10)
plt.xticks(rotation = 90)
plt.yticks(fontsize=10)
plt.xlabel("Estado", fontsize=10)
```

```python
locale_ei = engage_table.groupby(by=["locale"])['engagement_index'].mean()
labels = locale_ei.index  # x ticks
sizes = locale_ei.values

# Crear grafico circular
plt.figure(figsize = (7,7),dpi=100)
explode = [0.1,0,0,0]
colors = ['#2D87BB','#64C2A6','#AADEA7', "#E6F69D"]
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=False,textprops={'fontsize': 14},pctdistance=0.85,startangle=90, colors=colors)
#dibujar circulo
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Indice de Participación Medio por Región',color = 'Black',fontsize = 12)
```

```python
# Haciendo un Bar Plot
bh_ei = engage_table.groupby(by=["pct_black/hispanic"])['engagement_index'].mean()
plt.figure(figsize=(8,4),dpi=100)
plt.title('Porcentaje de Negros/Hispanos con Respecto al Indice Medio de Participacion')
sns.barplot(x=bh_ei.index,y=bh_ei.values, palette = "viridis",edgecolor='black')
plt.ylabel("Indice Medio de Participacion", fontsize=10)
plt.xticks(rotation = 90)
plt.yticks(fontsize=10)
plt.xlabel("Porcentaje de Negros/Hispanos", fontsize=10)
```
