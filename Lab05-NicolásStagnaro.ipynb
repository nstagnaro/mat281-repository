El **Iris dataset** es un conjunto de datos que contine una  muestras de tres especies de Iris (Iris setosa, Iris virginica e Iris versicolor). Se midió cuatro rasgos de cada muestra: el largo y ancho del sépalo y pétalo, en centímetros.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
# librerias
 
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', 500)  # Ver más columnas de los dataframes


# Ver gráficos de matplotlib en jupyter notebook/lab
%matplotlib inline
```

```python
# cargar datos
data = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_05/data/iris_contaminados.csv")
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

data.head() 
```

### Bases del experimento

Lo primero es identificar las variables que influyen en el estudio y la naturaleza de esta.

* **species**: 
    * Descripción: Nombre de la especie de Iris. 
    * Tipo de dato: *string*
    * Limitantes: solo existen tres tipos (setosa, virginia y versicolor).
* **sepalLength**: 
    * Descripción: largo del sépalo. 
    * Tipo de dato: *float*. 
    * Limitantes: los valores se encuentran entre 4.0 y 7.0 cm.
* **sepalWidth**: 
    * Descripción: ancho del sépalo. 
    * Tipo de dato: *float*. 
    * Limitantes: los valores se encuentran entre 2.0 y 4.5 cm.
* **petalLength**: 
    * Descripción: largo del pétalo. 
    * Tipo de dato: *float*. 
    * Limitantes: los valores se encuentran entre 1.0 y 7.0 cm.
* **petalWidth**: 
    * Descripción: ancho del pépalo. 
    * Tipo de dato: *float*. 
    * Limitantes: los valores se encuentran entre 0.1 y 2.5 cm.

Su objetivo es realizar un correcto **E.D.A.**, para esto debe seguir las siguientes intrucciones:

1. Realizar un conteo de elementos de la columna **species** y corregir según su criterio. Reemplace por "default" los valores `nan`.

```python
data['species'] = data['species'].str.lower() #Dejamos las valores de la columna species en minúscula
data['species'] = data['species'].str.strip() #Sacamos los espacios al principio y al final
data['species'].fillna("default",inplace = True) #Los valores de nan quedan reemplazados por default

cant_especies=len(data['species'].unique()) #Contamos la cantidad de elementos en la columna 'species'
print("La cantidad de elementos en la columna 'species' es:",cant_especies)
```

2. Realizar un gráfico de box-plot sobre el largo y ancho de los petalos y sépalos. Reemplace por **0** los valores nan.

```python
data['sepal_length'] = data['sepal_length']. fillna(0.0) #Donde haya un nan ponemos un 0
data['sepal_width'] = data['sepal_width']. fillna(0.0)
data['petal_length'] = data['petal_length']. fillna(0.0)
data['petal_width'] = data['petal_width']. fillna(0.0)

sns.boxplot(data=data)
plt.title('Tamaño y anchura de petalos y sepalos') 
plt.ylabel('Tamaño [cm]') #y al eje y
plt.show() #y lo mostramos
```

3. Anteriormente se define un rango de valores válidos para los valores del largo y ancho de los petalos y sépalos. Agregue una columna denominada **label**  que identifique cuál de estos valores esta fuera del rango de valores válidos.

```python
Label = []
for k in range(len(data)): #Ponemos una condición que verifique que todos los rangos se están cumpliendo. De no cumplirse solo uno, está fuera de rango.
  if (data['sepal_length'][k]>7.0 or data['sepal_length'][k]<4.0) or (data['sepal_width'][k]>4.5 or data['sepal_length'][k]<2.0) or (data['petal_length'][k]>7.0 or data['petal_length'][k]<1.0) or (data['petal_width'][k]>2.5 or data['petal_width'][k]<0.1):
    Label.append("Fail") #Algún parámetro no está en el rango
  else:
    Label.append("Pass") #Todos los parámetros están en los rangos

data["label"] = Label
data
```

4. Realice un gráfico de *sepal_length* vs *petal_length* y otro de *sepal_width* vs *petal_width* categorizados por la etiqueta **label**. Concluya sus resultados.

```python
plt.figure(figsize=(10, 6)) #Asignamos el tamaño
sns.scatterplot(x='sepal_length', y='petal_length', hue='label', data=data)
plt.title('sepal_length vs petal_length')
plt.show()
```

Conclusión: Hay pocos datos fuera de rango en el gráfico, y los que están fuera de rango suelen ser valores de orden superior. Además, los datos fuera de rango se acumulan en la parte superior derecha (solo un valor se acumula muy a la izquierda).

```python
plt.figure(figsize=(10, 6)) #Asignamos el tamaño
sns.scatterplot(x='sepal_width', y='petal_width', hue='label', data=data)
plt.title('sepal_width vs petal_width')
plt.show()# FIXME
```

Conclusión: Hay pocos datos fuera de rango en el gráfico, y los que están fuera de rango suelen ser valores superiores. También, los datos fuera de rango no están muy alejados entre sí.

5. Filtre los datos válidos y realice un gráfico de *sepal_length* vs *petal_length* categorizados por la etiqueta **species**.

```python
sns.set(rc={'figure.figsize':(10,8)})
df_specie=data.loc[data['label']=='Pass']
sns.scatterplot(x='sepal_length',y='petal_length',data=df_specie,hue='species')
plt.title('sepal_length vs petal_length')
plt.show()
```
