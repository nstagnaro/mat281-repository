EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# datos
x = df['month_number'].unique()
y = df['total_profit'].unique()

# grafico lineal

# tamano del grafico
fig = plt.figure(figsize=(8, 5))

# graficar
plt.plot(x, y, label='Beneficio Total', color="black", linewidth=1)
plt.legend()
plt.title("Beneficio Total vs Mes")
plt.xlabel('Mes')
plt.ylabel('Beneficio Total')
plt.show()
```

```python
# grafico de dispersion
fig = plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.7, label="Beneficio Total")
plt.title("Beneficio Total vs Mes")
plt.xlabel('Mes')
plt.ylabel('Beneficio Total')
plt.legend(loc='upper left')
plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
df_aux = df.drop(['month_number', 'total_units', 'total_profit'], axis = 1) #Filtramos las columnas que no nos sirven
fig = plt.figure(figsize=(8, 6))
for col in df_aux.columns: #Usamos un ciclo for para plotear cada curva pedida
    plt.plot(df['month_number'], df[col], label=col)


plt.legend()
plt.title('Ventas por producto')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.ylim(0, 15000)
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
facecream = df['facecream'].unique()
facewash = df['facewash'].unique()
labels = df['month_number'].unique()


ind = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8,6))
rects1 = ax.bar(ind - width/2, facecream, width, label='facecrame')
rects2 = ax.bar(ind + width/2, facewash, width, label='facewash')

ax.set_xlabel('Mes')
ax.set_ylabel('Ventas Totales')
ax.set_title('Gráfico de Barras Agrupado')
ax.set_xticks(ind)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
df_aux = df.drop(['month_number', 'total_units', 'total_profit'], axis = 1) #Filtramos las columnas que no nos sirven en este caso
fig = plt.figure(figsize=(8, 6))
sns.boxplot(data=df_aux)
plt.ylabel("Ventas Totales")
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular (de barra)

```python
df_aux2 = df_aux.sum() #Sumamos las ventas de cada producto en el año
fig = plt.figure(figsize=(8, 6))
plt.pie(df_aux2, labels=df_aux.columns, autopct='%1.1f%%') #Hacemos el grafico de barras
plt.title("Ventas Totales en el año")
plt.show()
```
